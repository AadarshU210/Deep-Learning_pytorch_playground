Basic math operations in Tensor


```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# Addition
tensor_a + tensor_b
```

```python
# Addition longhand
torch.add(tensor_a, tensor_b)
```

```python
# subtract
tensor_a - tensor_b
```

```python
torch.sub(tensor_a , tensor_b)
```

```python
# multiplication
tensor_a * tensor_b
```

```python
torch.mul(tensor_a, tensor_b)
```

```python
# Division
tensor_a / tensor_b
```

```python
torch.div(tensor_a, tensor_b)
```

```python
# Remainder Modulus
tensor_b % tensor_a
```

```python
torch.remainder(tensor_b, tensor_a)
```

```python
# Exponent / pow

tensor_a ** tensor_b
```

```python
torch.pow(tensor_a, tensor_b)
```
